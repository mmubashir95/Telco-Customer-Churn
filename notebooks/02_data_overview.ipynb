## Missing Values


```python
missing_values = df.isnull().sum().sort_values(ascending=False)
missing_values = missing_values[missing_values > 0]

if missing_values.empty:
    print("No missing values found.")
else:
    display(missing_values)
```

```python
df.duplicated().sum()
```

## Numerical And Categorical Columns


```python
numerical_columns = df.select_dtypes(include=["number"]).columns.tolist()
categorical_columns = df.select_dtypes(exclude=["number"]).columns.tolist()

{
    "numerical_columns": numerical_columns,
    "categorical_columns": categorical_columns,
}
```

## Target Distribution


```python
churn_counts = df["Churn"].value_counts()
churn_ratio = df["Churn"].value_counts(normalize=True).mul(100).round(2)

display(churn_counts)
display(churn_ratio)
```

## Target Distribution Ratio


```python
print("Churn is moderately imbalanced, so accuracy alone may be misleading.")
```

## Initial Observations

- The notebook relies on `load_raw_data()` and `validate_raw_data()` so the overview stage uses the same logic as the rest of the project.
- The dataset contains customer-level churn records with a mix of numeric and categorical features.
- `customerID` is an identifier and should not be treated as a predictive numerical feature.
- `Churn` is the target column, so its class balance should be reviewed before modeling.
- `TotalCharges` is currently loaded as `object`, so it needs cleaning/conversion in the data cleaning stage.

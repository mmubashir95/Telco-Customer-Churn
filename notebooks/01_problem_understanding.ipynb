## Business Problem

Customer churn means a customer stops using the telecom service. For the business, churn directly affects recurring revenue, acquisition efficiency, and long-term customer lifetime value.

A prediction system helps the company identify customers with high churn risk early enough to intervene with retention actions such as plan adjustments, support outreach, or targeted offers.


## Machine Learning Framing

| Item | Value |
| --- | --- |
| Problem Type | Binary Classification |
| Target | `Churn` |
| Positive Class | `Yes` |
| Negative Class | `No` |
| Unit of Analysis | One customer |

The goal is to estimate the probability that a customer will churn based on demographics, account setup, subscribed services, and billing behavior.

## Target Snapshot

Before modeling, it is important to understand how common churn is. This helps set expectations around class imbalance and metric choice.


```python
target_distribution = df['Churn'].value_counts()
target_share = df['Churn'].value_counts(normalize=True).mul(100).round(2)

print("Target distribution:")
display(target_distribution)

print("\nTarget share (%):")
display(target_share)
```

## Feature Categories

The available inputs naturally fall into a few groups:

- Customer profile: `gender`, `SeniorCitizen`, `Partner`, `Dependents`
- Account history: `tenure`, `Contract`, `PaperlessBilling`, `PaymentMethod`
- Service usage: `PhoneService`, `MultipleLines`, `InternetService`, `OnlineSecurity`, `OnlineBackup`, `DeviceProtection`, `TechSupport`, `StreamingTV`, `StreamingMovies`
- Charges: `MonthlyCharges`, `TotalCharges`

`customerID` is an identifier, not a predictive feature.


## Success Criteria

A useful churn model should do more than score well mathematically. It should support a business action.

Practical success criteria for this project:

- Identify a high-risk customer segment with meaningfully higher churn than the baseline rate.
- Maintain enough recall to catch customers who are likely to churn.
- Keep precision reasonable so retention efforts are not wasted on too many false positives.
- Produce interpretable drivers of churn so the results can inform business decisions.

Since missing churners is costly, recall and ROC-AUC will be especially important during model evaluation. Precision, F1-score, and confusion matrix review will also help assess whether the model is actionable for retention use.


## Assumptions And Risks

- The dataset is treated as a snapshot of historical customer states and churn outcomes.
- The cost of missing a likely churner may be higher than the cost of flagging a customer who would stay.
- Some fields may require cleaning or encoding before modeling, especially `TotalCharges`.
- If churn is imbalanced, accuracy alone will be a weak metric.
- Business usefulness depends on whether interventions can actually be applied to the flagged customers.
